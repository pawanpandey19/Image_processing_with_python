```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading and viewing the image from directory
ronaldo_image = plt.imread('ronaldo.jpeg')
plt.imshow(ronaldo_image)
plt.title('Image of Ronaldo')
plt.axis('off')
plt.show()
```

```python
# Converting the image into grayscale
from skimage import color
gray_image = color.rgb2gray(ronaldo_image)
plt.imshow(gray_image,cmap = 'gray')
plt.title('Grayscale image of Ronaldo')
plt.axis('off')
plt.show()
```
